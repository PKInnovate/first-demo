### Matrix Multiplication
Two main ways of multiplying tensors in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules for matrix multilpllication:
1. The **inner dimentions** must match.
* (3,2) @ (3, 2) won't work
* (2, 3) @ (3,2) will work
* (3, 2) @ (2, 3) will work
* (3, 3) @ (3, 3) will work

2. The resulting matrix has the shape of the **Outer dimentions**:
* (2,3) @ (3,2) -> (2,2)
* (3,2) @ (2, 3) -> (3,3)


```python
# Element wise multiplication
tensor * tensor
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

```python
%%timeit
torch.matmul(tensor, tensor)
```

## Reshaping , staccking, squeezing and unsqueezing the tensors

* **Reshaping** - reshapes an input tensor to a defined shape.
* **View** - Return a view of an input tnesor of certain shape but keep the same memoryas the original tensor.
* **Stacking** - Combine multiple tensors on top of each other (vstack) or side-by-side(hstack).
* **Squeeze** - revoes all '1' dimention from the tensor.
* **Unsqueeze** - add a '1' diemsion to target tensor.
* **Permute** - Return a view of the input with dimensions ermuted(swaped) in a certain way.


```python
# Lets create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped
x_reshaped, x_reshaped.shape
```

```python
 # Change the view
 z = x.view(1, 9)
 z, z.shape, z.size()
```

```python
# Chinging z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensor on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked, x_stacked.shape, x.size()
```

```python
# torch.squeeze() -> removes all single dimensions from a target tensor
x_reshaped, x_reshaped.shape
```

```python
x_reshaped.squeeze(), x_reshaped.squeeze().shape
```

```python
# torch.unsquueeze() -> adds a singe dimentiion to a target tensor at a specipic dim
x_squeezed = x_reshaped.squeeze()
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed, x_unsqueezed.shape
```

```python
# torch.permute -> rearranges the dimentsions of the target tensor in a speified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, color_channels]


# permute the original tensor to rearrange the azis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```
